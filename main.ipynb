```python
import math
from typing import Callable
```

```python
def f(x: float) -> float:
    return 0.25 * (x - 2.0) + 0.1 * math.sin(x)

def df(x: float, h: float) -> float:
    return (f(x + h) - f(x)) / h
```

```python
def bisseccao(
    f: Callable[[float], float],
    a: float,
    b: float,
    tolerancia: float
    ) -> tuple[float, int]:
    if f(a) * f(b) >= 0.0:
        raise ValueError("A função deve ter sinais diferentes em a e b.")

    i: int = 0

    while ((b - a) / 2.0) > tolerancia:
        i += 1
        c: float = (a + b) / 2.0

        if f(c) == 0.0:
            return c, i

        elif f(a) * f(c) < 0.0:
            b = c

        else:
            a = c

    return (a + b) / 2.0, i
```

```python
def posicao_falsa(
    f: Callable[[float], float],
    a: float,
    b: float,
    tolerancia: float
    ) -> tuple[float, int]:
    if f(a) * f(b) >= 0.0:
        raise ValueError("A função deve ter sinais diferentes em a e b.")

    i: int = 0

    while True:
        i += 1
        c: float = (a * f(b) - b * f(a)) / (f(b) - f(a))

        if abs(f(c)) <= tolerancia:
            return c, i

        if f(a) * f(c) < 0.0:
            b = c

        else:
            a = c
```

```python
def newton_raphson(
    f: Callable[[float], float],
    df: Callable[[float, float], float],
    x0: float,
    tolerancia: float,
    h: float = 1e-6,
    df_min: float = 1e-12,
    i_max: int = 100,
    ) -> tuple[float, int]:
    for i in range(1, i_max + 1):

        if abs(f(x0)) <= tolerancia:
            return x0, i

        if abs(df(x0, h)) < df_min:
            raise ValueError("Derivada muito próxima de zero.")

        x1: float = x0 - f(x0) / df(x0, h)

        if abs(x1 - x0) <= tolerancia:
            return x1, i

        x0 = x1

    raise ValueError("Não convergiu dentro do número máximo de iterações.")
```

```python
def secante(
    f: Callable[[float], float],
    x0: float,
    x1: float,
    tolerancia: float,
    d_min: float = 1e-12,
    i_max: int = 100,
    ) -> tuple[float, int]:
    for i in range(1, i_max + 1):
        if abs(f(x1) - f(x0)) < d_min:
            raise ValueError("Denominador muito pequeno.")

        x2: float = (f(x1) * x0 - f(x0) * x1) / (f(x1) - f(x0))
        f2: float = f(x2)

        if abs(f(x2)) <= tolerancia or abs(x2 - x1) <= tolerancia:
            return x2, i

        x0 = x1
        x1 = x2

    raise ValueError("Não convergiu dentro do número máximo de iterações.")
```

```python
tolerancia: float = 1e-6

r_bisseccao, i_bisseccao = bisseccao(f, 1.0, 2.0, tolerancia)
r_posicao_falsa, i_posicao_falsa = posicao_falsa(f, 1.0, 2.0, tolerancia)
r_newton, i_newton = newton_raphson(f, df, 1.5, tolerancia)
r_secante, i_secante = secante(f, 1.0, 2.0, tolerancia)

print(f"Bissecção      -> raiz: {r_bisseccao} | iterações: {i_bisseccao}")
print(f"Posição Falsa  -> raiz: {r_posicao_falsa} | iterações: {i_posicao_falsa}")
print(f"Newton-Raphson -> raiz: {r_newton} | iterações: {i_newton}")
print(f"Secante        -> raiz: {r_secante} | iterações: {i_secante}")
```
