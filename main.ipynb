```python
import math
from typing import Callable
```

```python
def f(x: float) -> float:
    return 0.25 * (x - 2.0) + 0.1 * math.sin(x)
```

**Bissecção**
- Definir $(a, b)$;
- Verificar se há solução em $(a, b)$;
- Partir $(a, b)$ na metade; $c = \frac{(b + a)}{2}$
- Verificar se $c$ é solução ou se a condição de parada é satisfeita;
- Definir novo $a$ ou novo $b$...

```python
def bissection(f: Callable[[float], float], a: float, b: float, tolerance: float) -> tuple[float, int]:
    if f(a) * f(b) >= 0.0:
        raise ValueError("A função deve ter sinais diferentes em a e b.")

    count: int = 0

    while ((b - a) / 2.0) > tolerance:
        count += 1
        c: float = (a + b) / 2.0

        if f(c) == 0.0:
            return c, count

        elif f(a) * f(c) < 0.0:
            b = c

        else:
            a = c

    return (a + b) / 2.0, count
```

```python
def false_position(f: Callable[[float], float], a: float, b: float, tolerance: float) -> tuple[float, int]:
    if f(a) * f(b) >= 0.0:
        raise ValueError("A função deve ter sinais diferentes em a e b.")

    count: int = 0

    while True:
        count += 1
        c: float = (a * f(b) - b * f(a)) / (f(b) - f(a))

        if abs(f(c)) <= tolerance:
            return c, count

        if f(a) * f(c) < 0.0:
            b = c

        else:
            a = c
```

```python
print(f"Bissecção:     {bissection(f, 1, 2, 1e-6)}")
print(f"Posição Falsa: {false_position(f, 1, 2, 1e-6)}")
```
